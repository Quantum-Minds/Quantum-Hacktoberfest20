# Quantum Binary Coded Decimal to Excess 3

```python
from qiskit import Aer, QuantumCircuit, execute
from qiskit.visualization import plot_histogram
```

```python
get_bin = lambda x, n: format(x, 'b').zfill(n)
```

```python
for i in range(0,10,1):
    circuit = QuantumCircuit(17,4)
    binary_no = get_bin(i, 4)
    if binary_no[-1] == '1':
        circuit.x(3)
    if binary_no[-2] == '1':
        circuit.x(2)
    if binary_no[-3] == '1':
        circuit.x(1)
    if binary_no[-4] == '1':
        circuit.x(0)

    # circuit.x([1,2,3])
    circuit.barrier()

    #####################################################
    # D
    circuit.x(3)
    circuit.barrier()
    circuit.measure(3,0)
    circuit.barrier()
    circuit.x(3)
    circuit.barrier()

    #####################################################
    # CD + C`D`
    circuit.ccx(2,3,14)
    circuit.swap(14,3)
    circuit.barrier()

    circuit.x([2,3])
    circuit.ccx(2,3,15)
    circuit.swap(15,3)
    circuit.barrier()

    circuit.x([14,15,16])
    circuit.ccx(14,15,16)
    circuit.x([14,15])
    circuit.swap(16,15)
    circuit.barrier()

    circuit.measure(15, 1)
    circuit.barrier()
    circuit.x([2,3])
    circuit.barrier()
    #####################################################

    # B AND D
    circuit.ccx(1,3,4)
    circuit.swap(4,3)
    circuit.barrier()

    # B AND C
    circuit.ccx(1,2,5)
    circuit.swap(5,2)
    circuit.barrier()

    # B AND D or B AND C
    circuit.x([2,3,6])
    circuit.ccx(2,3,6)
    circuit.x([2,3])
    circuit.swap(6,3)
    circuit.barrier()

    circuit.x([0,3,7])
    circuit.ccx(0,3,7)
    circuit.x([0,3])
    circuit.swap(7,3)
    circuit.barrier()

    circuit.measure(3, 3)
    circuit.barrier()

    #####################################################

    # NOT B AND D
    circuit.x(1)
    circuit.ccx(1,4,8)
    circuit.swap(8,4)
    circuit.barrier()

    # NOT B AND C
    circuit.ccx(1,5,9)
    circuit.swap(9,5)
    circuit.barrier()

    # B AND NOT C AND NOT D
    circuit.x(1)
    circuit.x(9)
    circuit.ccx(1,9,10)
    circuit.swap(10,9)
    circuit.barrier()

    circuit.x(8)
    circuit.ccx(8,9,11)
    circuit.swap(11,9)
    circuit.barrier()

    # NOT B AND D OR NOT B AND C OR B AND NOT C AND NOT D

    circuit.x([4,5,12])
    circuit.ccx(4,5,12)
    circuit.x([4,5])
    circuit.swap(12,5)
    circuit.barrier()

    circuit.x([5,9,13])
    circuit.ccx(5,9,13)
    circuit.x([5,9])
    circuit.swap(13,9)
    circuit.barrier()

    circuit.measure(9, 2)
    circuit.barrier()
    circ
    counts = execute(circuit, Aer.get_backend("qasm_simulator")).result().get_counts()
    print("{}\t--->\t{}\n".format(binary_no, max(counts)))
```
